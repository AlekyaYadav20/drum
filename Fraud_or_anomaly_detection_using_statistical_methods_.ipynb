```python
import pandas as pd
import numpy as np

# Sample transaction amounts
data = {'amount': [200, 220, 210, 215, 205, 5000]}
df = pd.DataFrame(data)

mean = df['amount'].mean()
std = df['amount'].std()

df['z_score'] = (df['amount'] - mean) / std

# Flag fraud
df['fraud'] = df['z_score'].abs() > 3

print(df)

Q1 = df['amount'].quantile(0.25)
Q3 = df['amount'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df['fraud'] = (df['amount'] < lower_bound) | (df['amount'] > upper_bound)

print(df)
threshold = mean + 3 * std
df['fraud'] = df['amount'] > threshold

print(df)
from scipy.stats import norm

df['probability'] = norm.pdf(df['amount'], mean, std)

# Low probability â†’ fraud
df['fraud'] = df['probability'] < 0.0001

print(df)
df['moving_avg'] = df['amount'].rolling(window=3).mean()
df['deviation'] = abs(df['amount'] - df['moving_avg'])

df['fraud'] = df['deviation'] > 2000




```

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("/content/nibss_fraud_dataset.csv")

# Display first rows
print(df.head())
amount_col = 'amount'   # change if your column name is different
mean = df[amount_col].mean()
std = df[amount_col].std()

df['z_score'] = (df[amount_col] - mean) / std

# Fraud condition
df['fraud_zscore'] = df['z_score'].abs() > 3

print(df[['z_score', 'fraud_zscore']].head())
Q1 = df[amount_col].quantile(0.25)
Q3 = df[amount_col].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df['fraud_iqr'] = (df[amount_col] < lower) | (df[amount_col] > upper)

print(df[['fraud_iqr']].head())
threshold = mean + 3 * std

df['fraud_std'] = df[amount_col] > threshold
from scipy.stats import norm

df['probability'] = norm.pdf(df[amount_col], mean, std)

df['fraud_prob'] = df['probability'] < 0.0001
df['final_fraud'] = (
    df['fraud_zscore'] |
    df['fraud_iqr'] |
    df['fraud_std'] |
    df['fraud_prob']
)

print(df[['final_fraud']].value_counts())
df.to_csv("fraud_detection_results.csv", index=False)
print("Fraud detection completed and saved.")
```
